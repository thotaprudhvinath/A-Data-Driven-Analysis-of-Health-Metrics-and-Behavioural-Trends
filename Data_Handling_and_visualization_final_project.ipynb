```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
body_health = pd.read_csv('health_dataset.csv')
```

```python
body_health
```

```python
body_health['date'] = pd.to_datetime(body_health['date'], errors='coerce')
```

```python
body_health = body_health.drop_duplicates()
```

```python
body_health.dropna()
```

```python
body_health.describe()
```

```python
 # Pie Chart: Distribution of Activity Levels
activity_counts = body_health['bool_of_active'].value_counts()
plt.figure(figsize=(8, 8))
activity_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('Set2', len(activity_counts))
)
plt.title('Activity Level Distribution')
plt.ylabel('')
plt.legend(labels=activity_counts.index, title='Activity Levels', loc='best')
plt.tight_layout()
plt.savefig('activity_level_distribution.png')
plt.show()
```

```python
# 2. Box Plot: Hours of Sleep by Mood
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=body_health,
    x='mood',
    y='hours_of_sleep',
    palette='coolwarm'
)
plt.title('Hours of Sleep by Mood')
plt.xlabel('Mood')
plt.ylabel('Hours of Sleep')
plt.tight_layout()
plt.savefig('hours_of_sleep_by_mood.png')
plt.show()
```

```python
# 3. Line Chart: Calories Burned Over Time with Trend Line
# Suppress warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=body_health,
    x='date',
    y='calories_burned',
    label='Calories Burned',
    color='blue',
    marker='o'
)
plt.title('Calories Burned Over Time')
plt.xlabel('Date')
plt.ylabel('Calories Burned')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('calories_burned_trend.png')
plt.show()
```

```python
# Suppress warnings
warnings.filterwarnings('ignore')

# Select a subset of numerical features for pairplot
pairplot_features = ['step_count', 'calories_burned', 'hours_of_sleep', 'weight_kg','mood']
data_for_pairplot = body_health[pairplot_features].dropna()

# Create a pairplot
sns.pairplot(
    data=data_for_pairplot,
    hue='mood',
    palette='coolwarm',
    kind='kde',  # Kernel Density Estimation
    diag_kind='kde',  # KDE for diagonal
    height=2.5
)

plt.suptitle('Pairwise Relationships and KDE by Mood', y=1.02)
plt.tight_layout()
plt.savefig('pairplot_with_kde.png')
plt.show()
```
